```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

<h1 style='color:blue'>LMPLOT</h1>

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df)
sns.lmplot(x='PetalLengthCm',y='PetalWidthCm',data=df)
```

<h1 style='color:red'>RESIPLOT</h1>

```python
sns.residplot(x=df['SepalLengthCm'], y=df['SepalWidthCm'], color='red')
```

<h1 style='color:yellow'>SCATTERPLOT</h1>

```python
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'],label='iris',color='red',marker='o')
```

<H1 style='color:purple;font:algerian;text-align:center'>REGPLOT</H1>

```python
sns.regplot(x= df['SepalLengthCm'],y=df['SepalWidthCm'],data='df',color='red',label='Order 1',order=1)
sns.regplot(x= df['SepalLengthCm'],y=df['SepalWidthCm'],data='df',color='blue',label='Order 2',order=2)
sns.regplot(x= df['SepalLengthCm'],y=df['SepalWidthCm'],data='df',color='yellow',label='Order 3',order=3)
```

<H1 style='color:lightgreen;font:algerian;text-align:center'>LINEAR REGRESSION BETWEEN THE VARIABLE(HUE)</H1>

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df,hue='Species',palette='Set1') 
```

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df,hue='Species',palette='Set2') 
```

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df,hue='Species',palette='Set3') 
```

<H1 style='color:PINK;font:algerian;text-align:center'>LINEAR REGRESSION BETWEEN THE VARIABLES (ROW-WISE)</H1>

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df,row='Species')
```

<H1 style='color:lime;font-family:algerian;text-align:center'><i>STRIPPLOT</i></H1>

```python
plt.subplot(2,2,1)
sns.stripplot(x='Species',y='SepalLengthCm',data=df)
plt.subplot(2,2,2)
sns.stripplot(x='Species',y='SepalWidthCm',data=df,jitter=True,size=4)
```

<H1 style='color:Green;font-family:cursive;text-align:center'>SWARMPLOT</H1>

```python
sns.swarmplot(x='Species',y='SepalWidthCm',data=df)
sns.swarmplot(x='Species',y='SepalLengthCm',data=df)
```

<H1 style='color:grey;font-family:courier;text-align:center'>VIOLINPLOT</H1>

```python
sns.violinplot(x='Species',y='SepalWidthCm',hue="Species",data=df,palette='Set1')
```

<STRONG><H1 style='color:maroon;font-family:courier;text-align:center'>JOINTPLOT</H1></STRONG>

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=df)
```

<STRONG><H1 style='color:lime-yellow;font-family:courier;text-align:center'>JOINTPLOT(PAIRWISE DISTRIBUTION)</H1></STRONG>

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=df,kind='resid')
```

<H1 style='color:brown;font-family:courier;text-align:center'>PAIRPLOT</H1>

```python
sns.pairplot(df)
```

<h1 style='color:lime;text-align:center'>BOXPLOT</h1>

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=df)
```
